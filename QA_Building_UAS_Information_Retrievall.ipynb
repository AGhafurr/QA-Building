Tokenisasi Kolom 'Catatan Amar'

```python
def count_tokens (text: str) -> int:
    return len(tokenizer.encode(text))
```

```python
if 'Catatan Amar' in df.columns:
  df['tokens'] = df['Catatan Amar'].apply(count_tokens)

df
```

Cleaning 'Catatan Amar'

```python
def clean_text(text):
    if isinstance(text, str):
        text = re.sub(r'[^\w\s]', ' ', text)  # Remove punctuation
        text = re.sub(r'\s+', ' ', text)  # Remove extra whitespace
        text = text.strip() # Remove leading/trailing whitespace
        text = text.lower() # Convert to lowercase
        return text
    else:
        return text

if 'Catatan Amar' in df.columns:
    df['Catatan Amar'] = df['Catatan Amar'].apply(clean_text)
```

## Get Questions

```python
def get_questions(context):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "Anda Merupakan asisten saya dalam mengarungi dunia."},
            {"role": "user", "content": f"Buatkan saya 5 pertanyaan berdasarkan text ini: \n\nText: {context}\n\nQuestion: \n1."}
        ],
    )
    # Access the content using .choices[0].message.content
    questions = response.choices[0].message.content.strip()
    return questions

df['questions'] = df['context'].apply(get_questions)
df.head()
```

## Get Answers

```python
def get_answers(context, question):
    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "system", "content": "Anda Merupakan asisten saya dalam mengarungi dunia."},
                {"role": "user", "content": f"Berikan jawaban untuk pertanyaan ini berdasarkan informasi ini: {context} \n\nQuestion: {question}"}
            ],
        )

        # Ambil konten menggunakan .choices[0].message.content
        answer = response.choices[0].message.content.strip()
        return answer
    except Exception as e:
        print(f"Error encountered for question: {question}. Error: {e}")
        return None  # Anda bisa mengembalikan nilai default lain jika perlu

# Menggunakan apply untuk setiap baris dalam DataFrame
df['answers'] = df.apply(lambda row: get_answers(row['context'], row['questions']), axis=1)

# Tampilkan hasil
df.head()
```

Proses pemisahan QA menjadi perkolom

```python
# Prepare a new DataFrame for split segments
df_split = pd.DataFrame(columns=['context', 'questions', 'answers'])

# Iterate and split/align context, question, and answer segments
for c, q, a in zip(df['context'], df['questions'], df['answers']):
    c_segments = str(c).split('\n')  # Ensure context is treated as a string
    q_segments = str(q).split('\n') # Ensure questions are treated as strings
    a_segments = str(a).split('\n\n')  # Ensure answers are treated as strings

    # Determine the maximum number of segments across context, questions, and answers
    max_segments = max(len(c_segments), len(q_segments), len(a_segments))

    # Loop through segments and append them to the new DataFrame
    for i in range(max_segments):
        #context = c_segments[i] if i < len(c_segments) else ''
        context = c_segments[i % len(c_segments)] if c_segments else None
        question = q_segments[i] if i < len(q_segments) else ''
        answer = a_segments[i] if i < len(a_segments) else ''
        df_split = pd.concat([df_split, pd.DataFrame({'context': [context],
                                                      'questions': [question],
                                                      'answers': [answer]})], ignore_index=True)

df_split
```

Embedding kolom Answer

```python
def embedding_answer(context):
    try:
        response = client.embeddings.create(
            model="text-embedding-3-small",
            input=context,
            encoding_format="float"
        )
        return response.data[0].embedding
    except Exception as e:
        print(f"An error occurred: {e}")

df_split['embedding_answer'] = df_split.answers.apply(embedding_answer)
```

Proses pengujian QA Fine-tuning

```python
from scipy.spatial.distance import cosine
import pandas as pd
import numpy as np

def rank_distances(questions_embedding, answer_embeddings_list):
    distances = []
    for i, embedding in enumerate(answer_embeddings_list):
        # Ensure embeddings are properly parsed or handled if they're strings
        if isinstance(embedding, str):
            try:
                # Convert the string representation into a list safely
                embedding = eval(embedding)  # alternative to ast.literal_eval with stricter control
            except (ValueError, SyntaxError):
                raise ValueError(f"Invalid embedding format at index {i}: {embedding}")

        # Ensure the embedding is in proper numerical format
        if isinstance(embedding, list):
            embedding = np.array(embedding)

        # Calculate cosine distance
        distance = cosine(questions_embedding, embedding)
        distances.append((i, distance))

    return sorted(distances, key=lambda x: x[1])
i = 12
question_embedding = train_df["Embedding_Q"][i]
ranked_results = rank_distances(question_embedding, train_df["Embedding_A"])

# Retrieve the top 5 results from ranked_results with respective ranks and answers
print("Pertanyaan: ", train_df["questions"][i])
top_5_results = [
    {"No": idx + 1, "Rank": rank, "Answer": df["answers"][i]}
    for idx, (i, rank) in enumerate(ranked_results[:5])
]

# Create a DataFrame for display
top_5_df = pd.DataFrame(top_5_results)
top_5_df
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np

# Fungsi untuk menghitung Precision@k
def precision_at_k(ranked_results, relevant_indexes, k=5):
    relevant_found = 0
    for idx, (i, _) in enumerate(ranked_results[:k]):
        if i in relevant_indexes:
            relevant_found += 1
    return relevant_found / k

# Fungsi untuk menghitung Recall@k
def recall_at_k(ranked_results, relevant_indexes, k=5):
    relevant_found = 0
    for idx, (i, _) in enumerate(ranked_results[:k]):
        if i in relevant_indexes:
            relevant_found += 1
    return relevant_found / len(relevant_indexes) if len(relevant_indexes) > 0 else 0

# Fungsi untuk menghitung F1 Score@k
def f1_at_k(ranked_results, relevant_indexes, k=5):
    relevant_found = [1 if i in relevant_indexes else 0 for i, _ in ranked_results[:k]]
    precision = np.sum(relevant_found) / k
    recall = np.sum(relevant_found) / len(relevant_indexes) if len(relevant_indexes) > 0 else 0
    if precision + recall > 0:
        return 2 * (precision * recall) / (precision + recall)
    else:
        return 0

# Fungsi untuk menghitung MRR
def mean_reciprocal_rank(ranked_results, relevant_indexes):
    for idx, (i, _) in enumerate(ranked_results):
        if i in relevant_indexes:
            return 1 / (idx + 1)  # MRR menghitung posisi relevan pertama
    return 0  # Jika tidak ada jawaban relevan

# Fungsi untuk evaluasi matrix
def evaluate_at_k(ranked_results, relevant_indexes, k=5):
    precision = precision_at_k(ranked_results, relevant_indexes, k)
    recall = recall_at_k(ranked_results, relevant_indexes, k)
    f1 = f1_at_k(ranked_results, relevant_indexes, k)
    mrr = mean_reciprocal_rank(ranked_results, relevant_indexes)

    return {
        "Precision@k": precision,
        "Recall@k": recall,
        "F1@k": f1,
        "MRR": mrr
    }

# Contoh penggunaan
relevant_indexes = [0, 2]  # Indeks jawaban relevan untuk pertanyaan ini
i = 12
question_embedding = train_df["Embedding_Q"][i]
ranked_results = rank_distances(question_embedding, train_df["Embedding_A"])

# Evaluasi dengan top 5 hasil
evaluation_metrics = evaluate_at_k(ranked_results, relevant_indexes, k=5)
print("Evaluation Metrics: ", evaluation_metrics)
```
